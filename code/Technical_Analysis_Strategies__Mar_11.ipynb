```python
# 数据预处理
spy_growth['date'] = pd.to_datetime(spy_growth['date'])
spy_value['date'] = pd.to_datetime(spy_value['date'])
spy_growth['return_label'] = spy_growth['return'].apply(lambda x: '+' if x > 0 else '-')
spy_value['return_label'] = spy_value['return'].apply(lambda x: '+' if x > 0 else '-')

# 构建特征
def create_label_sequence(data):
    sequences = []
    for i in range(len(data)):
        if i >= 9:  # 确保有足够的数据来创建10天的序列
            sequence = ''.join(data['return_label'][i-9:i+1])
            sequences.append(sequence)
        else:
            sequences.append(None)  # 对于序列开始的部分，填充None
    return sequences

spy_growth['feature'] = create_label_sequence(spy_growth)
spy_value['feature'] = create_label_sequence(spy_value)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 初始化投资金额
initial_investment = 100

# 初始化存储投资值的列表
buy_hold_growth_values = [initial_investment]
buy_hold_value_values = [initial_investment]
ew_values = [initial_investment]  # 等权重策略的投资价值列表

# 初始化日期列表
buy_hold_dates = [pd.to_datetime('2001-01-01')]

# 遍历每个日期，计算投资价值
for date in pd.date_range(start='2001-01-01', end='2023-12-31'):
    if date in spy_growth['date'].values and date in spy_value['date'].values:
        growth_rate = spy_growth.loc[spy_growth['date'] == date, 'return'].values[0]
        value_rate = spy_value.loc[spy_value['date'] == date, 'return'].values[0]

        # 计算'buy and hold'策略的投资价值
        current_value_growth = buy_hold_growth_values[-1] * (1 + growth_rate)
        current_value_value = buy_hold_value_values[-1] * (1 + value_rate)

        # 更新'buy and hold'策略的价值和日期
        buy_hold_growth_values.append(current_value_growth)
        buy_hold_value_values.append(current_value_value)
        buy_hold_dates.append(date)

        # 计算等权重策略的投资价值
        current_value_ew = ew_values[-1] * (0.5 * (1 + growth_rate) + 0.5 * (1 + value_rate))
        ew_values.append(current_value_ew)

```

```python

# 绘图比较策略
plt.figure(figsize=(14, 7))

# 绘制'buy and hold spy growth'策略
plt.plot(buy_hold_dates, buy_hold_growth_values, label='Buy and Hold SPY Growth')

# 绘制'buy and hold spy value'策略
plt.plot(buy_hold_dates, buy_hold_value_values, label='Buy and Hold SPY Value')
plt.plot(valid_dates, sp500_values, label='S&P 500 Investment')
# 绘制等权重策略
plt.plot(buy_hold_dates, ew_values, label='Equal Weight (EW) Strategy', linestyle='--', color='purple')
# 设置图例，并指定图例字体大小和位置
plt.legend(fontsize='x-large', ncol=3)  # 增加字体大小并将图例移动到上方中间，ncol用于控制图例的列数
plt.title('Strategy Performance Comparison (EW)')
plt.xlabel('Date')
plt.ylabel('Investment Value ($)')
plt.show()
```

```python
import pandas as pd
import numpy as np

# 加载数据
sp500 = pd.read_csv('/content/spy.csv')
spy_growth = pd.read_csv('/content/spyg.csv')
spy_value = pd.read_csv('/content/spyv.csv')

# 日期格式转换
sp500['date'] = pd.to_datetime(sp500['date'])
spy_growth['date'] = pd.to_datetime(spy_growth['date'])
spy_value['date'] = pd.to_datetime(spy_value['date'])

# 计算每年的波动性函数，使用实际天数
def calculate_annual_volatility(returns_df):
    returns_df['Year'] = returns_df['date'].dt.year
    annual_volatility = returns_df.groupby('Year')['return'].std() * (returns_df.groupby('Year')['return'].count() ** 0.5)
    return annual_volatility

# 计算每年的波动率
annual_volatility_spy = calculate_annual_volatility(sp500)
annual_volatility_growth = calculate_annual_volatility(spy_growth)
annual_volatility_value = calculate_annual_volatility(spy_value)

# 整理数据并保存到CSV
annual_volatility_df = pd.DataFrame({
    'Year': annual_volatility_spy.index,
    'SPY': annual_volatility_spy.values,
    'Growth': annual_volatility_growth.values,
    'Value': annual_volatility_value.values
})

annual_volatility_df.to_csv('/content/annual_volatility_comparison.csv', index=False)

print("Annualized volatility data for SPY, Growth, and Value strategies have been saved to CSV file.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 假设 sp500, spy_growth, 和 spy_value 数据已经被加载
# 你应该已经有了sp500, spy_growth, 和 spy_value DataFrame，每个DataFrame包含'date'和'return'列

# 确保日期列的类型正确
sp500['date'] = pd.to_datetime(sp500['date'])
spy_growth['date'] = pd.to_datetime(spy_growth['date'])
spy_value['date'] = pd.to_datetime(spy_value['date'])

# 合并数据以便于计算相关性
merged_data = pd.merge(sp500, spy_growth, on='date', suffixes=('_spy', '_growth'))
merged_data = pd.merge(merged_data, spy_value, on='date')
merged_data.rename(columns={'return': 'return_value'}, inplace=True)

# 添加年份列以便分组
merged_data['Year'] = merged_data['date'].dt.year

# 初始化存储年度相关性的字典
annual_correlation_spy_growth = {}
annual_correlation_spy_value = {}
annual_correlation_value_growth = {}

# 计算每年的相关性
for year, group in merged_data.groupby('Year'):
    correlation_matrix = group[['return_spy', 'return_growth', 'return_value']].corr()
    annual_correlation_spy_growth[year] = correlation_matrix.loc['return_spy', 'return_growth']
    annual_correlation_spy_value[year] = correlation_matrix.loc['return_spy', 'return_value']
    annual_correlation_value_growth[year] = correlation_matrix.loc['return_value', 'return_growth']

# 绘制相关性图表
def plot_correlation(correlation_data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(list(correlation_data.keys()), list(correlation_data.values()), marker='o')
    plt.title(title)
    plt.xlabel('Year')
    plt.ylabel('Correlation')
    plt.grid(True)
    plt.show()

# 绘制SPY与Growth的年度相关性
plot_correlation(annual_correlation_spy_growth, 'Annual Correlation between SPY and Growth')

# 绘制SPY与Value的年度相关性
plot_correlation(annual_correlation_spy_value, 'Annual Correlation between SPY and Value')

# 绘制Value与Growth的年度相关性
plot_correlation(annual_correlation_value_growth, 'Annual Correlation between Value and Growth')
```

```python
import matplotlib.pyplot as plt

def plot_combined_correlation(correlation_data_dict, start_year=2006, end_year=2023):
    plt.figure(figsize=(14, 7))

    # 定义颜色映射，为不同的数据序列指定颜色
    color_map = {
        'SPY and Growth': 'blue',
        'SPY and Value': 'green',
        'Value and Growth': 'red'
    }

    # 对每个相关性数据序列进行绘制
    for label, correlation_data in correlation_data_dict.items():
        # 筛选指定年份范围的数据
        filtered_keys = [year for year in correlation_data.keys() if start_year <= year <= end_year]
        filtered_values = [correlation_data[year] for year in filtered_keys]
        plt.plot(filtered_keys, filtered_values, marker='o', linestyle='-', label=label, color=color_map[label])

    plt.title('Annual Correlation Comparison')
    plt.xlabel('Year')
    plt.ylabel('Correlation')
    # 明确设置x轴的刻度标签为2006-2023年
    plt.xticks(range(start_year, end_year + 1))
    plt.grid(True)
    plt.legend()
    plt.show()

# 假设 annual_correlation_spy_growth, annual_correlation_spy_value, 和 annual_correlation_value_growth
# 已经按照之前的步骤计算出来了

# 将三个相关性数据序列合并到一个字典中
correlation_data_dict = {
    'SPY and Growth': annual_correlation_spy_growth,
    'SPY and Value': annual_correlation_spy_value,
    'Value and Growth': annual_correlation_value_growth
}

# 绘制合并的相关性图表
plot_combined_correlation(correlation_data_dict)
```
